```python
import pandas as pd
%matplotlib inline
```

## Analysis with 5 Samples

```python
raw=pd.read_csv("avg_5_samplev2.csv")
df_raw=pd.DataFrame(raw)
df_5_sample=df_raw.iloc[:,10:19]
# df_5_sample.head()
indx=["plt1","plt2","plt3","plt4","plt5","plt6","plt7","plt8","plt9"]
df_5_sample.columns=indx
df_5_sample.to_csv("sample_5.csv")
df_5_sample.head()
```

```python
stats5=df_5_sample.describe()
print(type(stats5))
stats5
```

```python
std=stats5.loc['std',:]
avg=stats5.loc['mean',:]
CV5=pd.DataFrame(std/avg)
CV=CV5.T
new_stats=pd.concat([stats5,CV])
last = new_stats.index[-1]
new_stats.rename(index={last: 'CV'})
```

```python
df_5_sample['plt4'].plot()
df_5_sample['plt3'].plot()
```
